```python
import pandas as pd
import numpy as np

data = {
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107],
    "Name": ["Amit", "Neha", "Rohit", "Simran", "Pooja", "Rahul", "Neha"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [60000, 45000, 70000, 80000, np.nan, 65000, 72000],
    "Experience": [2, 3, 5, 7, 1, 4, 6]
}

df = pd.DataFrame(data)
df
```

```python
df[(df['Department']=='IT') &( df['Salary']>60000)]
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df['Salary']=df['Salary'].fillna(df.groupby('Department')['Salary'].transform('mean'))
```
