```python
import seaborn as sns

df=sns.load_dataset('tips')
```

```python
df
```

```python
def smoking(val):
  if val=='Yes':
    return 1
  else:
    return 0

df['smoking']=df.smoker.apply(smoking)

df
```

```python
df.smoking.value_counts()
```

```python
def tips(val):
  if val>5:
    return 'rich'
  else:
    return 'poor'

df['tips']=df['tip'].apply(tips)

df
```
